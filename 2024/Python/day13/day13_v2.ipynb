````python
rules = input[:-1].split("\n\n")
print(rules)
````

````python
for i in range(len(rules)):
    rules[i] = rules[i].split("\n")
    for r in range(len(rules[i])):
        rules[i][r] = rules[i][r].split(": ")
print(rules)
````

````python
def button_rule(rule):
    buttonLabel = rule[0][-1]
    xy = rule[1].split(", ")
    x = int(xy[0][2:])
    y = int(xy[1][2:])
    return (buttonLabel, x, y)

for r in rules:
    r[0] = button_rule(r[0])
    r[1] = button_rule(r[1])
    r[2] = r[2][1].split(", ")
    r[2] = (int(r[2][0][2:]), int(r[2][1][2:]))
    print(r)   
````

````python
rules = list(map(lambda r: (r[0], r[1], (int(r[2][0]) + 10000000000000, int(r[2][1]) + 10000000000000)), rules))
print(rules)
````

````python
from math import isclose
solution = 0

for r in rules:
    ax = r[0][1]
    ay = r[0][2]
    bx = r[1][1]
    by = r[1][2]
    cx = r[2][0]
    cy = r[2][1]
    b1 = (ax * cy) - (ay * cx)
    b2 = (ax * by) - (bx * ay)
    b = b1 / b2
    a = (cx - (bx * b)) / ax
    xx = (a * ax) + (b * bx)
    yy = (a * ay) + (b * by)
    if(xx == cx and yy == cy):
        if a > 0 and b > 0 and isclose(a, int(a)) and isclose(b, int(b)):
            print(f"xx == cx and yy == cy: {xx} == {cx} and {yy} == {cy}")
            print(f"a = {a}, b = {b}")
            solution += (3 * int(a) + int(b))
````
